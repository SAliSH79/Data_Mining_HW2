```python
data = pd.read_csv("housePrice.csv")
data.head() #Showing some rows of the dataset
```

## Preparation

```python
data #General overview about dataset
```

```python
data_2 = data.dropna() #Ridding of empty cells
data_2
```

```python
data_2.Price.describe() # Fiding the required parameters
```

```python
#Now, we will create a mapping function (salary_stats) and use the DataFrame.map()
#function to create a new column from an existing column
def price_level(value):
  if value < 1.415e+9: 
    return "cheap"
  if 1.415e+9 <= value < 2.88e+9:
    return "undermean"
  if 2.88e+9 <= value < 6e+9:
    return "Uppermean"
  elif 6e+9 <= value <= 9.24e+9:
    return "expensive"
```

```python
data_2['Price_Level'] = data_2['Price'].map(price_level)
#Using DataFrame.map() function to create new column from existing column using a mapping function
display(data_2.head()) 
```

```python
data_2
```

##### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
#create instance of label encoder
lab = LabelEncoder()
```

```python
data_2['Parking'] = lab.fit_transform(data_2['Parking'])
data_2['Warehouse'] = lab.fit_transform(data_2['Warehouse'])
data_2['Elevator'] = lab.fit_transform(data_2['Elevator'])
data_2['Address'] = lab.fit_transform(data_2['Address'])
data_2
```

##### Normalization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
data_2['Area'] = scaler.fit_transform(data_2['Area'].values.reshape(-1,1))[:,0]
data_2['Room'] = scaler.fit_transform(data_2['Room'].values.reshape(-1,1))[:,0]
data_2['Parking'] = scaler.fit_transform(data_2['Parking'].values.reshape(-1,1))[:,0]
data_2['Warehouse'] = scaler.fit_transform(data_2['Warehouse'].values.reshape(-1,1))[:,0]
data_2['Elevator'] = scaler.fit_transform(data_2['Elevator'].values.reshape(-1,1))[:,0]
data_2['Address'] = scaler.fit_transform(data_2['Address'].values.reshape(-1,1))[:,0]
data_2['Price'] = scaler.fit_transform(data_2['Price'].values.reshape(-1,1))[:,0]
data_2
```

##### Train & Test spliting

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(data_2, test_size=0.2)
train
```
